# MNB Feature Weighting

```python
# the feature file is an edited copy of Weka's MultinomialNB output
import pandas as pd
import numpy as np
weka_output = pd.read_csv('mnb-20newsgroups.tsv', delimiter='\t')
features = weka_output['token'].values

alt_atheism_prob = weka_output['alt.atheism'].values
comp_graphics_prob = weka_output['comp.graphics'].values
comp_os_ms_windows_misc = weka_output['comp.os.ms-windows.misc'].values
comp_sys_ibm_pc_hardware = weka_output['comp.sys.ibm.pc.hardware'].values
comp_sys_mac_hardware = weka_output['comp.sys.mac.hardware'].values
comp_windows_x = weka_output['comp.windows.x'].values
misc_forsale = weka_output['misc.forsale'].values
rec_autos = weka_output['rec.autos'].values
rec_motorcycles = weka_output['rec.motorcycles'].values
rec_sport_baseball = weka_output['rec.sport.baseball'].values
rec_sport_hockey = weka_output['rec.sport.hockey'].values
sci_crypt = weka_output['sci.crypt'].values
sci_electronics = weka_output['sci.electronics'].values
sci_med = weka_output['sci.med'].values
sci_space = weka_output['sci.space'].values
soc_religion_christian = weka_output['soc.religion.christian'].values
talk_politics_guns = weka_output['talk.politics.guns'].values
talk_politics_mideast = weka_output['talk.politics.mideast'].values
talk_politics_misc = weka_output['talk.politics.misc'].values
talk_religion_misc = weka_output['talk.religion.misc'].values
```

```python
all_else_prob = []
for i in range(0, len(features)):
    all_sum = comp_graphics_prob[i] + comp_os_ms_windows_misc[i] + comp_sys_ibm_pc_hardware[i] + comp_sys_mac_hardware[i] + \
              comp_windows_x[i] + misc_forsale[i] + rec_autos[i] + rec_motorcycles[i] + rec_sport_baseball[i] + rec_sport_hockey[i] +\
              sci_crypt[i] + sci_electronics[i] + sci_med[i] + sci_space[i] + soc_religion_christian[i] + talk_politics_guns[i] +\
              talk_politics_mideast[i] + talk_politics_misc[i] + talk_religion_misc[i]
    all_else_prob.append(all_sum)
```

```python
print(features[0])
print(alt_atheism_prob[0])
print(all_else_prob[0])
```

```python
import math
log_ratios = []
for i in range(0, len(features)):
    if alt_atheism_prob[i] > 0 and all_else_prob[i] > 0:
        log_ratio = math.log(alt_atheism_prob[i]) - math.log(all_else_prob[i])
        log_ratios.append(log_ratio)
    else: 
        log_ratio = 0
        log_ratios.append(log_ratio)
print(log_ratios[0])
```

```python
features = pd.DataFrame(features)
log_ratios = pd.DataFrame(np.array(log_ratios))
feature_ranks = pd.concat([features, log_ratios], axis=1, ignore_index=True)
feature_ranks.columns = ['token', 'log_ratios']
feature_ranks = feature_ranks.sort_values('log_ratios')
feature_ranks[:-20]
```

```python
feature_ranks = sorted(zip(log_ratios, features))

# print the words with highest pos/neg conditional prob ratio / most positive words
#top_atheism_features = feature_ranks[-20:]
#print(top_atheism_features)
```
